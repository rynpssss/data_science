### 因子分析
- 目的
    - 取得データの背景に共通の要因が隠れていることを目的に行う　※データ間で相関関係があることが前提
- 因子負荷量の計算
	- 主成分と変数との相関が因子負荷量
- 共通性　＋　独自性　＝　１
- 解の求め方は、最尤法や最小二乗法の他に主因子法や主成分法がある
- https://norimune.net/705#:~:text=%E6%9C%80%E5%B0%A4%E6%B3%95%E3%81%A8%E3%81%84%E3%81%86%E3%81%AE,%E5%B0%A4%E6%B3%95%E3%81%A8%E3%81%84%E3%81%84%E3%81%BE%E3%81%99%E3%80%82
	- 最尤法
		- 最も使われている
		- 漸近一致性・漸近有効性・漸近正規性を持っている
		- データは正規分布でないといけない
	- 最小二乗法
		- 正規性を仮定しない
		- 最尤法との違いは、誤差の重み付け
			- 全ての変数の誤差を同じ重みで考える　→　共通性が低い項目の影響もウケる
				- 最尤法　→　共通性が小さい項目は重みが小さい
		- それっぽい答えを出す時には有効
- YouTube（アイシア）
	- 因子構造とデータ要約ができる
	- 次元削減
	- 解釈がむずい・・（回転が必要）
		- 理由
			- 両方の軸が増える　→　元々相関をとっているので当たり前（情報が増えない）
			- どちらかが増えて、どちらかが減る　→　良いことか分からない（それってどうゆうこと？）
			- →　どちらかが減るのは解釈が難しいから、どちらかは0でどちらが増えるとする
		- バリマックス回転後の因子負荷行列を見ることで、結果の解釈が容易になる
			- 何かが増えるだけになる
			- 因子分析に行列Rを作用させて、別の行列から解釈を行う
		- バリマックス回転（直交回転）
			- 分散の最大化
			- （因子負荷量）の二乗の因子内分散の和を最大化
			- 因子負荷量の絶対値が大きくなり、結果の解釈がしやすい
			- 0が多くなる
			- 因子が被らなくなる
			- 最強
				- 無相関
				- Rが正方行列で、R転置*R＝1
				- 直交行列を用いた回転では、分散の最大化が可能で無相関を保てる
		- プロマックス回転（斜行回転）
			- 互いに無相関ではない場合もあるので、その場合はこちら
			- プロマックス回転　＝　バリマックス回転　＋　プロクラステス回転　※よりバリマックス回転を強調させるために使う
			- プロクラテス回転
				- ちょっと無理矢理合わせる
				- ステップ４つ
					- １（仮説を用意）
					- ２（因子分析）
					- ３（無理矢理合わせる）
					- ４（まとめ）
				- 本物の因子不可行列を仮説に合うように回転する（普通の流れとは逆）
				- 分散を1にするために転置Tをできる
				- ※　色々な計算式があるので、ドキュメントの確認必須
			- プロマックス回転
				- ステップ
					- １（因子分析）
					- ２（バリマックス回転）
					- ３（プロクラテス回転）
						- 仮説行列をバリマックス回転の因子負荷行列の四乗を用いる
						- 四乗の理由
							- バリマックス回転によって、すでに因子負荷量の大小があった
							- 四乗によって、さらに強調を高める
			- バリマックスによって、程よく因子構造が明らかになった。しかし、直交回転の制約下で因子がわかりきれない場所があった
			- そこで、さらにプロマックス回転で因子構造をより強調するようにする
