# SVM

## 特徴

- 判別平面と最も近いデータの距離が最大になるように判別平面を求めて分類を行う方法
- ランダムフォレストとの使い分け
	- データ数が多い＆次元数が多いか？　※計算量の問題
		- Yes　→　ランダムフォレスト（求め方の都合上、データが多いほど精度が高まる）
		- No     →　SVM
	- 軸に意味を持っているか？
		- Yes　→　ランダムフォレスト
		- No　→　SVM
		- →　軸に意味を持っていて、高次元になるとどうやってもSVMで滑らかな超平面は作れない

## 機械学習図鑑
目的：めちゃくちゃざっくりと概要だけ知る<br>

SVMは、分類と回帰どちらにも使える

### 線形SVM
#### 概要
線形SVMはマージンの最大化という基準により、決定境界をデータから出来るだけ離れるように学習させること

#### アルゴリズム
線形SVMでの二値分類は、2次元に直線を引いて分類する考え方<br>
この時、決定境界から最も近いデータと決定境界との距離をマージンと呼ぶ

#### ソフトマージンとサポートベクトル
 - ハードマージン：マージンの内側に入り込まない分類　※通常は完全に分類出来ることがないので使えない
 - ソフトマージン：一部のデータはマージンの内側や決定境界よりも外側になることを許容する分類

ソフトマージンでは、学習データは以下3種類
 - 決定境界からマージンよりも離れているデータ：マージンの外側
 - 決定境界からマージンと同じ距離のデータ：マージン上
 - 決定境界にマージンよりも近い、もしくは誤分類されるデータ：マージンの内側
この時、「マージンの外側」はどうでもよい<br>
「マージン上」「マージンの内側」のデータを**サポートベクトル**と呼び、決定境界を決める重要なデータとなる

### サポートベクトルマシン（カーネル法）
#### 概要
線形SVMでは、良い決定境界を作れるが曲線の分類が苦手　→ カーネル法を用いる

#### アルゴリズム
**「データを別の特徴空間に飛ばして線形回帰を行う」**
学習データよりも高次元データがあり、学習データはそれぞれ高次元空間に対応している<br>
高次元空間の中では線形回帰が可能であり、実際の学習データは高次元空間からの射影と考える<br>

高次元空間を具体的に構成するのは困難だが、カーネル法のカーネル関数を利用する事で構築できる

#### 詳細
カーネル関数にはさまざまな種類があるため、状況によって使い分ける必要がある
 - 線形カーネル
 - シグモイドカーネル
 - 多項カーネル
 - RBFカーネル
