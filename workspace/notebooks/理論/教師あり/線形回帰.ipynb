# 線形回帰

- 推定方法の違い
    - 最尤推定とは？
        - データから本来の確率（尤度）を推定すること
        - 推定とは？　→　データから確率を導くこと
        - 尤度とは？　→　現象から真の確率を導く際に、現象からデータ最も大きな確率を尤度（最も、真の確率だろう値）
        - 尤度の最大を求めるのは、ちょっとめんどいので最大対数尤度を求める
        - 最大対数尤度の微分が０を計算
        - 最尤推定の問題点
            - 回数が少ない時の最尤推定はまともな答えが出ない。
    - ベイズ推定とは？
        - 出現すると予測される確率分布（事前分布）と実際の結果を元に事後分布を求め、最大値を推定する方法
        - 最尤推定との違い
            - 最尤推定は、結果のみを世の中全ての現象とし、その中で最も確率が高い尤度を求める方法
            - ベイズ推定は、世の中で一般的に起こるであろう確率分布を事前分布として予測し、結果と合わせて最大確率を求める方法
        - 事前分布とは？
        
        
- 導出方法の違い
    - なぜ最小「二乗」法なのか？
        - 計算がシンプル（最小二乗法をy = ax + bとすると、a,bは分散と共分散を使えば求められる）
        - 絶対値は大変
        - BLUEだから（最も適切な推定量だから）
            - BLUEの特性
                - 線形性：線形モデルであること
                - 一致性：nを大きくすると、真の値に一致
                - 普遍性：推定量の期待値が真の値に一致
                - 効率性：分散が最小
            - ガウス・マルコフの定理
                - 均一分散
                - 共分散０
                - 誤差項と説明変数の相関０
                - つまり、最小二乗推定量はBLUEであるという証明
    - 最小二乗法は、誤差を正規分布を仮定した時に利用される（そのため、回帰診断を行うことが重要）　※誤差が正規分布を仮定した最尤推定と同義
        - 正規分布でなくても使えるが、誤差の中にモデル化できる要素が含まれている　＝　回帰式が適切なモデルではないため
    - 最尤推定は、正規分布を仮定しないGLMにて使用可能
        - ポアソン分布や二項分布なども仮定
    - 最尤推定と最小二乗法の違い
        - 平均値の求め方が異なる
        - 最尤推定は、尤度が最大となる平均値を探す
        - 最小二乗法は、誤差が最小となる平均値を探す
        - 最尤推定は、モデルの拡張が容易。
            - 線形でなくても、指数関数的な関係でも対応可能
            - また、誤差が正規分布ではなく、ポアソン分布や二項分布でも可能
- 決定係数
    - 全体の変動から、残差変動を引いた値
    - つまり、どの程度の変動を説明出来ているのかの指標
- 結論：線形回帰にて、誤差が正規分布の時、最小二乗法と最尤推定は結果が同じ
		- 誤差が正規分布ではない場合には最小二乗法を使用したモデリングはやめた方が良い。
		- 線形回帰は見た目で2次元の時に可視化しやすい
		- 最小二乗法による回帰は、現象の数式理解が容易

## 久保本
### 線形回帰まとめ

===================

#### 1章
線形モデル（LM）：ばらつきは正規分布を仮定<br>
一般化線形モデル（GLM）：いやいや、なんでもばらつきを正規分布に仮定するのはおかしい！　→ もっと色んな分布を扱いたい<br>
一般化線形混合モデル（GLMM）：「個体差」「場所差」などを表現するモデル<br>

===================

#### 2章
目的変数の図示によるデータの確率分布を検討する　→ っぽい分布が見つかればOK<br>
 ※個体ごとに同一の確率分布に従っているが、なんらかの理由で誤差が発生している<br>
 
データにばらつきがある　＝　統計モデリングで事象を表現できそうと考えることが一歩目<br>
　　人には見えない真の分布がある　→ そこから、サンプリングされてデータを取得　→ サンプリングデータから真の分布を推定<br>
  
現実のデータは確率分布で表現できるほど単純ではない？<br>
→ いや、取得データの環境の違いや取得条件を見直せば、ある程度は確率分布で扱う事が可能なはず！！！<br>

===================

#### 3章　GLM　※なんでも正規分布はおかしいだろww

初めにやる事：色んな図にして、データをよく見る（EDA）　<br>


GLMを特徴づける概念<br>
 - 線形予測子　：　y = x1 + x2 + x3 の{x1, x2, x3}のようにxが線形結合になっている
 - リンク関数　：　「y = 線形予測子」となっている場合、左辺の関数をリンク関数と呼ぶ
 
対数リンク関数は掛け算で表されるのに対し、恒等リンク関数は足し算で表される

等分散の正規分布かつ「リンク関数なし」の恒等リンク関数とすると、GLMではなくLM

===================

#### 4章　GLMのモデル選択（ AICの説明　）

上記に記載。
良いモデルとは平均対数尤度が高いモデルと定義して、<br>
最大対数尤度とパラメータ数から推定する

===================

#### 5章　尤度比検定

AICとの違い
 - 帰無仮説が棄却されたからといって良いモデルとは限らない

検定の注意点
 - 棄却したモデルがどういったモデルかまでは説明できない
 - 帰無仮説に選ぶネストしたモデルが、真のモデルとして検定が行われてしまう
 
結論として、上記の理由から検定を頻繁に使う機会は少なそう。

===================

#### 6章　GLMの応用

最初に、GLMと変数変換した直線回帰は全く異なる結果となる。

モデル選択は、AICで選ぶのが良いだろう

交互作用項（A * B）は、のちに解釈が難しくなるのでむやみに入れない方が無難

オフセット項　：　線形予測子の中で、係数がついていない値 ※線形予測子にオフセット項という「げた」をはかせている<br>
→ 異なる条件で観測されたデータに対して、GLMを適応する場合に有効

正規分布<br>
最小二乗法と最尤推定の関係　：　標準偏差が一定である正規分布のパラメータの最尤推定　＝　最小二乗法の推定

===================

#### 7章　GLMM

GLMM　：　人間が測定できない・測定できなかった個体差を含んだモデル<br>
データのばらつきはポアソン分布・個体のばらつきは正規分布等の複数の確率分布を部品とできる統計モデル<br>
例　ロジスティック関数<br>
GLM  → logit(y) = b1 + b2xi<br>
GLMM → logit(y) = b1 + b2xi + ri　※riは個体差<br>

GLMMの特徴
- 個体差を表すパラメータが何か確率分布に従っているかを仮定するところ
- 個体差だけの反復だと場所差がある場合は、個体差と場所差の判定ができない為、両方を反復させる必要がある

===================

#### 8章　MCMC

MCMCによってサンプリングされた定常分布は、尤度に比例する確率分布を表す

頻度主義　：　パラメータは一意な数　※確率分布ではない<br>
ベイズ統計　：　パラメータは確率分布

ベイズ統計とMCMC<br>
事後確率　＝　MCMCの確率分布　＝　尤度×事前分布（ベイズ統計）

===================

#### 9章　GLMのベイズ化

ベイズ統計の手順

    1.事前分布の決定
        無情報として、-∞ ~ ∞ までをとる一様分布としたいが積分して1となるために、範囲を-10^9 ~ 10^9　くらいにする
    2.事後確率の推定
        MCMCによるサンプリング　※初期値依存による除外、期間は自ら確認しながら対処
        収束診断と^Rの情報を確認する　→ ^Rの値が1に近ければ、収束している　※1.1を超えるとダメ！
    3.複数パラメータによる同時分布の削除
        積分して、求めたいパラメータのみの周辺事後分布にする
        
ギブスサンプリングの場合<br>
パラメータを同時に推定することは出来ないので、多変量の場合は1変量にして他を定数とみなしてからMCMCによるパラメータの更新を行う<br>
全パラメータを更新させるまでが1巡。これをMCMCのループ分回す。
     
        
ギブスサンプリング ※メトロポリス法との相違点
 - メトロポリス　→ 新しい値の候補をあげ、それに変化するかどうかを決める
 - ギブスサンプリング　→ 新しい値の確率分布を作り、その確率分布のランダムサンプルを新しい値とする　<br>
     ※新しい確率分布とは、求めたいパラメータのみの一変量確率分布（他パラメータは定数扱い）
     
**AICのようなモデルの良さをベイズ統計は比較できない**<br>
**クロスバリデーションで評価するのが最も良いっぽい・・・**

===================

#### 10章　GLMMのベイズ化（階層ベイズモデル）

GLMでは無情報事前分布を使用してベイズ統計を設計したが、<br>
GLMMでは**階層事前分布**を使って階層ベイズモデルを設計する

過分散が起きている等で、確率分布では表現できな場合に有効

階層ベイズモデルとは？<br>
→ 階層（複数段）のベイズモデルを推定すること<br>
　　※ 事前分布のパラメータにさらに事前分布のパラメータを設定している為、事前分布の事前分布は超事前分布と呼ばれ、階層になっている
  
無情報事前分布と階層事前分布の選び方<br>
パラメータがデータ全体のどの範囲を説明しているのかに依存する
 - 範囲が大域　かつパラメータ数は少数　→ 無情報事前分布
 - 範囲が局所　かつパラメータ数は多数　→ 階層事前分布

===================

#### 11章　空間構造のある階層ベイズモデル

===================

#### まとめ

LMとGLMの違い　→ データを正規分布と仮定するか、他の確率分布と仮定するか<br>
GLMとGLMMの違い　→ 1つの確率分布を仮定するか、複数の確率分布を仮定するか<br>

GLMM<br>
 - 見えない潜在変数を取り入れたGLM（分布を混ぜるという考え方）
 - 「同じ個体や場所から何度もサンプリングしているか？」　→ あれば、GLMM
 - 「個体差や場所差を認識できるデータがあるか？」→ 無ければ、GLMM
 - 個体差を表す特徴が、確率分布に従っていると仮定する
 - 混合の意味は、固定効果とランダム効果が混合している為
 
GLMM求め方
 - 「計測できていない変数を積分して消す」or「MCMC」

===================

総括
 - データが全て正規分布を仮定するのはおかしい　→ GLMを使うべき
 - データが1つの確率分布のみから生成されると考えるのはおかしい　→ GLMMを使うべき
 - LM・GLMは最尤推定にて、パラメータの推定を行う
 - GLMMはMCMCサンプリングにてパラメータを推定する
     - 以下2種類があるが、ギブスサンプリングの方が共役事前分布を仮定できるので計算が楽になる
     - メトロポリス法にて、繰り返しサンプル生成する
     - ギブスサンプリングにて、事前分布を決めてサンプル生成する
 - GLMMで難しいのは、階層ベイズモデルを取るので自分でモデル構築をしなければならない
     - 事前分布を無情報事前分布と階層事前分布に分けて設定する
     - MCMCにてサンプル生成した事後分布を確認して、データの当てはまりを見る
     - AICのようなモデルの良さを決める指標がないので、クロスバリデーションにて自分で判断する必要がある
